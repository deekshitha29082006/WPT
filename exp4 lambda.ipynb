```python
import pandas as pd
import numpy as np
data={
    'First':['Aryan','Rohan','Riya','Yash','Siddini'],
    'Lastname':['Singh','Agarwal','Shah','Batia','Khan'],
    'Type':['Full_time employee','Intern','Full_time employee','part time employee','Full_time employee'],
    'Department':['Adminstration','Technical','Adminstration','Technical','Technical'],
    'Yoe':[2,3,5,7,6],
    'Salary':[20000,5000,10000,20000, 0]

    }
df=pd.DataFrame(data)
print(df)
table=pd.pivot_table(df,index=['Type','Department'],
values='Salary',aggfunc=['mean'])
print(table)
table=pd.pivot_table(df,index=['Type','Department'],
                       values='Salary',aggfunc=['sum','mean'])
print(table)
table2=pd.pivot_table(df,index=['Type'],
values='Salary',aggfunc=['count'])
print(table2)
table3=pd.pivot_table(df,index=['Type'],
values='Salary',aggfunc=['std'])
print(table3)
```
